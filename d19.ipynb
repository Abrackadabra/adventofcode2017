```python
pos = array([0, where(M[0, :] != ' ')[0][0]])

diri = 0

def get(x):
    if any(x >= array(M.shape)) or any(x < zeros(2)):
#         print('out', x)
        return ' '
    return M[tuple(x)]

def dirc(diri):
    return '|' if diri % 2 == 0 else '-'

counter = 0

while True:
# for i in range(100):
    c = get(pos)
    counter += 1
#     print(pos, c, diri)
    
    if c == ' ':
        print()
        print(f'end at {pos}')
        break
    
    if c == '|' or c == '-':
        pos += dirs[diri]
        continue
    
    if c not in ' -|+':
        print(c, end='')
        pos += dirs[diri]
        continue
    
    if c == '+':
        if get(pos + dirs[diri]) != ' ':
            pos += dirs[diri]
            continue
        
        posdiri = (diri + 1) % 4
        pospos = pos + dirs[posdiri]
        g = get(pospos)
        
#         print('?', posdiri, pospos, g)
        if g != ' ':
#             print(f'{diri} -> {posdiri}')
            diri = posdiri
            pos = pospos
            continue
            
        posdiri = (diri + 3) % 4
        pospos = pos + dirs[posdiri]
        g = get(pospos)
        
#         print('?', posdiri, pospos, g)
        if g != ' ':
#             print(f'{diri} -> {posdiri}')
            diri = posdiri
            pos = pospos
            continue
        
        print(':(')
        break
        
counter - 1
```
