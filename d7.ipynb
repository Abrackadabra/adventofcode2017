```python
weights = {}
edges = defaultdict(set)
parents = {}

for line in lines:
    parts = line.split(' -> ')
    name, weight = re.findall(r'^(\w+) \((\d+)\)', parts[0])[0]
    weight = int(weight)
    
    children = set() if len(parts) == 1 else parts[1].split(', ')
    edges[name] = children
    weights[name] = weight
    for c in children:
        parents[c] = name
```

```python
set(list(weights.keys())) - set(list(parents.keys()))
```

```python
def total_weight(name):
    t = 0
    for i in edges[name]:
        t += total_weight(i)
    return t + weights[name]
```

```python
for i in weights:
    cn = edges[i]
    c = [total_weight(j) for j in edges[i]]
    if c:
        mask = argsort(c)
        cn = array(cn)[mask]
        c = array(c)[mask]

        if c[0] != c[-1]:
            print(i, c)
            print(i, cn)
            print(i, [weights[j] for j in cn])
```
