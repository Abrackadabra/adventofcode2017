```python
def execute(p_id, commands, pos, values, q_in, q_out):
    global counter
    
    def what(s):
        try:
            return int(s)
        except:
            return values[s]
    
    parts = commands[pos].split()
    
    cmd = parts[0]
    a = parts[1]
    b = parts[2] if len(parts) >= 3 else '0'
    
    a_v = what(a)
    b_v = what(b)
    
#     print(f'{pos}: {cmd} {a}={a_v} {b}={b_v}', flush=True)
    
    if cmd == 'set':
        values[a] = b_v
    elif cmd == 'add':
        values[a] += b_v
    elif cmd == 'mul':
        values[a] *= b_v
    elif cmd == 'mod':
        values[a] %= b_v
    elif cmd == 'jgz':
        if a_v > 0:
            pos += b_v - 1
    elif cmd == 'snd':
        q_out.append(a_v)
        if p_id == 1:
            counter += 1
    elif cmd == 'rcv':
        if q_in:
            values[a] = q_in.popleft()
        else:
            pos -= 1
            
    return pos + 1
```
