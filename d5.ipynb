```python
import requests

cookies = {
    'session': '53616c7465645f5f465a94508e1037cf931dedaf44e298b1be500ec71305ff01432a6584b1fe922ea5bab725d96f17b1'
}

url = 'http://adventofcode.com/2017/day/5/input'

r = requests.get(url, cookies=cookies)

lines = [i.strip() for i in r.text.split('\n')][:-1]
```

```python
a = array([int(i) for i in lines])
```

```python
x = array(a)
t = 0
c = 0
while 0 <= t < len(x):
    tt = t + x[t]
    if x[t] >= 3:
        x[t] -= 1
    else:
        x[t] += 1
    c += 1
    t = tt
    
    if c % 100000 == 0:
        print(sum(x))
t, c
```
