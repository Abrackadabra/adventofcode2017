```python
def what(s):
    try:
        return int(s)
    except:
        return values[s]
    
mul_count = 0

def execute(command):
    global values
    global pos
    global mul_count
    
    if pos == 11:
        print('pos=11', values)
        
        if values['b'] % values['d'] == 0:
            z = values['b'] // values['d']
            if values['e'] <= z <= values['b']:
                values['f'] = 0
        values['e'] = values['b']
        pos = 20
        print('end   ', values)
        
        return
    
    parts = command.split()
    
    cmd = parts[0]
    a = parts[1]
    if len(parts) >= 3:
        b = parts[2]
    else:
        b = '0'
    
    a_v = what(a)
    b_v = what(b)
    
#     print(f'{pos}: {cmd} {a}={a_v} {b}={b_v}', flush=True)
    
    if cmd == 'set':
        values[a] = b_v
    elif cmd == 'sub':
        values[a] -= b_v
    elif cmd == 'mul':
        values[a] *= b_v
        mul_count += 1
    elif cmd == 'jnz':
        if a_v != 0:
            pos += b_v - 1
    else:
        print('wat', cmd)
            
    pos += 1
```
